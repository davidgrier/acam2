```python
%matplotlib inline
from acam2 import acoustic_camera
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
```

```python
a = acoustic_camera()
```

```python
scan_range = 60
y0 = 5
step = 0.5
d_8k_04 = a.scan(scan_range=scan_range, y0 = y0, step = step)
```

```python
d = d_6k_01
grid_x, grid_y = np.mgrid[-30:30:128j, y0 + 4:64:128j]
amp = griddata(d[:,0:2],d[:,2],(grid_x, grid_y), method='cubic')
phi = griddata(d[:,0:2],d[:,3],(grid_x,grid_y), method='cubic')
plt.subplot(121)
plt.imshow(amp.T)
plt.subplot(122)
plt.imshow(phi.T)
```
